```python
# Python 3
# Load in CSV containing emoji and coordinates
# Display the emojis on a map using Folium
# Note this uses sample date with random coordinates, emojis, place names and types
```

```python
import pandas as pd
import folium
from folium.features import DivIcon # for displaying emojis this worked best
```

```python
# load data
emoji_place_sample_df = pd.read_csv('sample_place_points_with_emojis.csv')
print(emoji_place_sample_df.shape)
emoji_place_sample_df.head()
```

```python
# clean data and get center coords

# if any null coordinates drop them first
emoji_places_df = emoji_place_sample_df.dropna()
print('rows of data remaining:', len(emoji_places_df))

# get center_lngX and center_latY
center_lngX = emoji_places_df['lng_X'].median()
center_latY = emoji_places_df['lat_Y'].median()
print('center coords lngX, latY:', center_lngX, center_latY)
```

```python
# make a map using folium

center_latY = 38.89
center_lngX = -77.03

map = folium.Map(location=[center_latY, center_lngX], zoom_start=14, control_scale = True,
                 tiles='http://services.arcgisonline.com/ArcGIS/rest/services/Canvas/World_Light_Gray_Base/MapServer/tile/{z}/{y}/{x}.png',
                 attr='ESRI')


for _, row in emoji_places_df.iterrows():
    folium.Marker(location=[row["lat_Y"], row["lng_X"]], icon=DivIcon(icon_size=(150,36), icon_anchor=(-1,-1),
        html='<div style="font-size: 12pt">{}</div>'.format(row['top_emoji'])), tooltip=row['sample_place_name']).add_to(map)
map

```
